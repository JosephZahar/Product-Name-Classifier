# Automatic Product Labelling using Python

```python
import pandas as pd
```

```python
# transform the csv into a dataframe in Python for manipulation
data = pd.read_csv('/Users/macbookpro/Desktop/XXXX/Product Labelling using ML/labels.csv')
```

```python
# import the product names and classify them into their categories
new_list = [] # create an empty list
for row in data.values.ravel(): # iterates through the value of each row of the dataframe
    if row[1] != '%':  # check if the second index of the string is a %
        new_list.append(row[1:-1])  # if not append the value to the new_list
new_list = new_list[:-9]   # remove the last 9 elements from the list
```

```python
category_1 = new_list[0:3] # the first three element of the list are the labels of the first hierarchy (Food, Drink, Unspecified)
category_2 = new_list[3:5] # the two next elements are the labels of the second hierarchy (GF, Not GF)
category_3 = new_list[5:7] # the two next elements are the labels of the third hierarchy (Vegan, Not Vegan)
category_4 = new_list[7:32] # the rest of the  elements are the labels of the fourth hierarchy
```

```python
new_list_1 = []  # creates a new list
for row in data.values.ravel(): # iterates through the value of each row of the dataframe
    if row[1] == '%': # check if the second index of the string is a %
        new_list_1.append(row[2:-2]) # if it is append the value to the new_list without the %
```

```python
Drinks = [] # new list that will contain all the drinks requirements
for i in new_list_1: # iterates through each element of new_list_1 created just above
    if i != 'CHICKEN': # if element not 'CHICKEN'
        Drinks.append(i) # Add it to the list Drinks
    else:
        break # Stop appending values to the list when we reach the string 'CHICKEN'
remove = ['BATTERED', 'TIGER BREAD', 'GINGER', 'STEAK'] # list of strings to remove from Drinks
for i in remove: # remove each element individually from the list
    Drinks.remove(i)
Drinks = Drinks[:-15] # remove the last 15 elements from the list
Food = new_list_1[new_list_1.index(Drinks[-1]) + 1:new_list_1.index('ROASTED') + 1] # retrieve all the strings from the last element of Drinks to 'ROASTED'
Food.remove('TEA') # remove TEA from list
Gluten_free = new_list_1[new_list_1.index(Food[-1]) + 4:new_list_1.index('(GF)') + 1] # retrieve all the strings from the last element of Food to 'GF'
Vegan = new_list_1[new_list_1.index(Gluten_free[-1]) + 6:new_list_1.index('(VG/') + 1] # retrieve all the strings from the last element of Gluten_free to 'GF'
Cider = new_list_1[new_list_1.index(Vegan[-1]) + 2:new_list_1.index(Vegan[-1]) + 17] # retrieve all the strings from the last element of Vegan to a specific index
Beer = new_list_1[new_list_1.index(Vegan[-1]) + 18:new_list_1.index(Vegan[-1]) + 78] # retrieve all the strings from a specific index to another index
remove = ['KHAO', 'BATTERED', 'SOLEIL', 'SOLI', 'SCHNITZEL', 'PEPPERONI', 'TIGER BREAD', 'SCHWEPPES']
for i in remove:
    Beer.remove(i) # remove these strings from the Beer list
Spirits = new_list_1[new_list_1.index(Vegan[-1]) + 83:new_list_1.index(Vegan[-1]) + 155] # retrieve all the strings from a specific index to another index
Cocktails = new_list_1[new_list_1.index(Vegan[-1]) + 156:new_list_1.index(Vegan[-1]) + 187] # retrieve all the strings from a specific index to another index
Wine = new_list_1[new_list_1.index(Vegan[-1]) + 188:new_list_1.index(Vegan[-1]) + 229] # retrieve all the strings from a specific index to another index
Coffee = new_list_1[new_list_1.index(Vegan[-1]) + 231:new_list_1.index(Vegan[-1]) + 244] # retrieve all the strings from a specific index to another index
Tea = ['TEA'] # define Tea
Soft_drinks = new_list_1[new_list_1.index(Vegan[-1]) + 245:new_list_1.index(Vegan[-1]) + 285] # retrieve all the strings from a specific index to another index
Other_drinks = new_list_1[new_list_1.index(Vegan[-1]) + 285:new_list_1.index(Vegan[-1]) + 291] # retrieve all the strings from a specific index to another index
Chicken = ['COQ AU VIN', 'CHICKEN', 'WINGS'] # define Chicken
Beef = new_list_1[new_list_1.index(Vegan[-1]) + 299:new_list_1.index(Vegan[-1]) + 303] # retrieve all the strings from a specific index to another index
Pork = ['PORK'] # define Pork
Other_meat = new_list_1[new_list_1.index(Vegan[-1]) + 306:new_list_1.index(Vegan[-1]) + 322] # retrieve all the strings from a specific index to another index
Pizza = new_list_1[new_list_1.index(Vegan[-1]) + 322:new_list_1.index(Vegan[-1]) + 326] # retrieve all the strings from a specific index to another index
Potato = new_list_1[new_list_1.index(Vegan[-1]) + 326:new_list_1.index(Vegan[-1]) + 330] # retrieve all the strings from a specific index to another index
Burger = new_list_1[new_list_1.index(Vegan[-1]) + 331:new_list_1.index(Vegan[-1]) + 333] # retrieve all the strings from a specific index to another index
Seafood = new_list_1[new_list_1.index(Vegan[-1]) + 334:new_list_1.index(Vegan[-1]) + 344] # retrieve all the strings from a specific index to another index
Pasta = new_list_1[new_list_1.index(Vegan[-1]) + 344:new_list_1.index(Vegan[-1]) + 352] # retrieve all the strings from a specific index to another index
Souces = new_list_1[new_list_1.index(Vegan[-1]) + 352:new_list_1.index(Vegan[-1]) + 359] # retrieve all the strings from a specific index to another index
Sandwich = new_list_1[new_list_1.index(Vegan[-1]) + 359:new_list_1.index(Vegan[-1]) + 365] # retrieve all the strings from a specific index to another index
Sandwich.remove('BANANA BREAD') # remove 'BANANA BREAD' from Sandwich
Breakfast = new_list_1[new_list_1.index(Vegan[-1]) + 369:new_list_1.index(Vegan[-1]) + 375] # retrieve all the strings from a specific index to another index
Breakfast.remove('TEA')  # remove 'TEA' from Breakfast
Salad = new_list_1[new_list_1.index(Vegan[-1]) + 375:new_list_1.index(Vegan[-1]) + 380] # retrieve all the strings from a specific index to another index
Soup = ['SOUP'] # define Soup
Other_food = new_list_1[new_list_1.index(Vegan[-1]) + 381:new_list_1.index(Vegan[-1]) + 405] # retrieve all the strings from a specific index to another index
Nuts = new_list_1[new_list_1.index(Vegan[-1]) + 405:new_list_1.index(Vegan[-1]) + 409] # retrieve all the strings from a specific index to another index
```

```python
def label_products(table): # need a dataframe as an input
    label_1 = [] # create an empty list that will contain the hierarchy 1 label of each products of the table
    for row in table.values.ravel(): # iterates through the values of each row of the table
        if any(drink in row.upper() for drink in Drinks): # Check if the product contain any of the strings that are in the Drink list
            label_1.append('Drink') # if yes label it as Drink
        elif any(food in row.upper() for food in Food): # if not check if the product contain any of the strings that are in the Food list
            label_1.append('Food') # if yes label it as Food
        else:
            label_1.append('Unspecified') # else label it as Unspecified

    label_2 = [] # create an empty list that will contain the hierarchy 2 label of each products of the table
    for row in table.values.ravel(): # iterates through the values of each row of the table
        if any(gf in row.upper() for gf in Gluten_free): # Check if the product contain any of the strings that are in the Gluten_free list
            label_2.append('Gluten Free') # if yes label it as GF
        else:
            label_2.append('Not Gluten Free') # else label it as not GF

    label_3 = [] # create an empty list that will contain the hierarchy 3 label of each products of the table
    for row in table.values.ravel(): # iterates through the values of each row of the table
        if any(vg in row.upper() for vg in Vegan): # Check if the product contain any of the strings that are in the Vegan list
            label_3.append('Vegan') # if yes label it as Vegan
        else:
            label_3.append('Not Vegan')  # else label it as not Vegan

    label_4 = [] # create an empty list that will contain the hierarchy 3 label of each products of the table
    for row in table.values.ravel(): # iterates through the values of each row of the table
        if any(i in row.upper() for i in Cider): # Check if the product contain any of the strings that are in the Cider list
            label_4.append('Cider')  # if yes label it as Cider
        elif any(i in row.upper() for i in Beer): # if not check if the product contain any of the strings that are in the Beer list
            label_4.append('Beer') # if yes label it as Beer
        elif any(i in row.upper() for i in Spirits): # if not check if the product contain any of the strings that are in the Spirits list
            label_4.append('Spirits') # if yes label it as Spirits
        elif any(i in row.upper() for i in Cocktails): # if not check if the product contain any of the strings that are in the Cocktails list
            label_4.append('Cocktails') # if yes label it as Cocktails
        elif any(i in row.upper() for i in Wine): # if not check if the product contain any of the strings that are in the Wine list
            label_4.append('Wine') # if yes label it as Wine
        elif any(i in row.upper() for i in Coffee): # if not check if the product contain any of the strings that are in the Coffee list
            label_4.append('Coffee') # if yes label it as Coffee
        elif any(i in row.upper() for i in Tea): # if not check if the product contain any of the strings that are in the Tea list
            label_4.append('Tea') # if yes label it as Tea
        elif any(i in row.upper() for i in Soft_drinks): # if not check if the product contain any of the strings that are in the Soft_drinks list
            label_4.append('Soft drinks') # if yes label it as Soft_drinks
        elif any(i in row.upper() for i in Other_drinks): # if not check if the product contain any of the strings that are in the Other_drinks list
            label_4.append('Other drinks') # if yes label it as Other_drinks
        elif any(i in row.upper() for i in Chicken): # if not check if the product contain any of the strings that are in the Chicken list
            label_4.append('Chicken') # if yes label it as Chicken
        elif any(i in row.upper() for i in Beef): # if not check if the product contain any of the strings that are in the Beef list
            label_4.append('Beef') # if yes label it as Beef
        elif any(i in row.upper() for i in Pork): # if not check if the product contain any of the strings that are in the Pork list
            label_4.append('Pork') # if yes label it as Pork
        elif any(i in row.upper() for i in Other_meat): # if not check if the product contain any of the strings that are in the Other_meat list
            label_4.append('Other meat') # if yes label it as Other_meat
        elif any(i in row.upper() for i in Pizza): # if not check if the product contain any of the strings that are in the Pizza list
            label_4.append('Pizza') # if yes label it as Pizza
        elif any(i in row.upper() for i in Potato): # if not check if the product contain any of the strings that are in the Potato list
            label_4.append('Potato') # if yes label it as Potato
        elif any(i in row.upper() for i in Burger): # if not check if the product contain any of the strings that are in the Burger list
            label_4.append('Burger') # if yes label it as Burger
        elif any(i in row.upper() for i in Seafood): # if not check if the product contain any of the strings that are in the Seafood list
            label_4.append('Seafood') # if yes label it as Seafood
        elif any(i in row.upper() for i in Pasta): # if not check if the product contain any of the strings that are in the Pasta list
            label_4.append('Pasta') # if yes label it as Pasta
        elif any(i in row.upper() for i in Souces): # if not check if the product contain any of the strings that are in the Souces list
            label_4.append('Souces') # if yes label it as Souces
        elif any(i in row.upper() for i in Sandwich): # if not check if the product contain any of the strings that are in the Sandwich listc
            label_4.append('Sandwich') # if yes label it as Sandwich
        elif any(i in row.upper() for i in Breakfast): # if not check if the product contain any of the strings that are in the Breakfast list
            label_4.append('Breakfast') # if yes label it as Breakfast
        elif any(i in row.upper() for i in Salad): # if not check if the product contain any of the strings that are in the Salad list
            label_4.append('Salad') # if yes label it as Salad
        elif any(i in row.upper() for i in Soup): # if not check if the product contain any of the strings that are in the Soup list
            label_4.append('Soup') # if yes label it as Soup
        elif any(i in row.upper() for i in Other_food): # if not check if the product contain any of the strings that are in the Other_food list
            label_4.append('Other food') # if yes label it as Other_food
        elif any(i in row.upper() for i in Nuts): # if not check if the product contain any of the strings that are in the Nuts list
            label_4.append('Nuts') # if yes label it as Nuts
        else:
            label_4.append('Unknown') # else label it as Unknown

    table.loc[:, 'hierarchy 1'] = label_1 # create a new column where each row's value is the label in the list label_1
    table.loc[:, 'hierarchy 2'] = label_2 # create a new column where each row's value is the label in the list label_2
    table.loc[:, 'hierarchy 3'] = label_3 # create a new column where each row's value is the label in the list label_3
    table.loc[:, 'hierarchy 4'] = label_4 # create a new column where each row's value is the label in the list label_4
    return table # return the dataframe as a csv and save it


data = pd.read_csv('/Users/macbookpro/Desktop/XXXX/Product Labelling using ML/Unlabelled Product Names.csv')
df = label_products(data)
df
```
